## 4/1 출석자 찾기

```python
for i in range(0, len(df)):
    if str(df.at[i, '4/1']).find('Attended') != -1:
        print(df.at[i, 'Name'])
```

## 4/1 지각자 찾기

```python
for i in range(0, len(df)):
    if str(df.at[i, '4/1']).find('Late') != -1:
        print(df.at[i, 'Name'])
```

## 4/2 결석자 찾기

```python
for i in range(0, len(df)):
    if str(df.at[i, '4/2']) == 'Absence':
        print(df.at[i, 'Name'])
```

## 4/2에 Joel이 배운 과목 이름 찾기

```python
name_str = str(input('학생 이름 : '))
name = -1

for i in range(0, len(df)):
    if df.at[i, 'Name'] == name_str:
        name = i

if name == -1:
    print('해당 학생은 우리 학원의 원생이 아닙니다.')
    exit()

date = str(input('날짜 : '))

print(name)
print(date)
print(str(df.at[name, date]))

if str(df.at[name, date]) == 'No' or str(df.at[name, date]) == 'Absence':
    print(name + '학생은 ' + date + '에 출석하지 않았습니다.')
else:
    start_index = str(df.at[name, date]).find('Subject') + 10
    end_index = str(df.at[name, date]).find('Teacher') - 1
    string = str(df.at[name, date])
    print('과목명 : ' + string[start_index:end_index])
```

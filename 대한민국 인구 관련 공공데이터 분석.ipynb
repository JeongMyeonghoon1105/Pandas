```python
import pandas as pd

df = pd.read_csv('Data/Population.csv')
```

```python
df = df[['행정구역', '2022년04월_계_총인구수']]
```

```python
df.columns = ['행정구역', '22년 4월 총 인구수']
```

```python
df['1인 연 100만원 기준 지급액 총액']
```

```python
df
```

```python
def remove_comma(x):
    return x.replace(',', '')

for i in range(0, len(df)):
    df.at[i, '22년 4월 총 인구수'] = remove_comma(df.at[i, '22년 4월 총 인구수'])
    df.at[i, '1인 연 100만원 기준 지급액 총액'] = int(float(df.at[i, '22년 4월 총 인구수'])) * 1000000
```
